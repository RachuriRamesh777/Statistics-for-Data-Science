Problem Statement

A city claims that the average weight of its people is 72 kg with a population standard deviation of 4 kg.

A random sample of 36 people is taken from the city, and the sample mean weight is found to be 74 kg.

Using a 95% confidence level, test whether the sample provides sufficient statistical evidence to reject the null hypothesis that the population mean weight is 72 kg.

The problem requires performing a hypothesis test (Z-test) to determine whether the observed sample mean significantly differs from the claimed population mean.

Population mean μ=72 kg

Population standard deviation σ=4 kg

Sample size n=36

Sample mean xˉ=74 kg

Confidence level = 95%

Significance level α=0.05


1. Step 1 – Define Hypotheses

We test whether the population mean differs from 72.

𝐻0:𝜇=72

𝐻𝑎:𝜇≠72

This is a two-tailed test.

2. Step 2 – Why Z-Test?

We use a Z-test because:

Population standard deviation σ is known

Sample size n = 36 ≥ 30

Therefore the Central Limit Theorem applies.

3. Step 3 – Z-Test Formula

𝑍=𝑥ˉ -𝜇/𝜎/ROOT n

Where

𝑥bar = sample mean

μ = population mean

σ = population standard deviation

n = sample size

4. Step 4 – Calculate Standard Error

SE= sigma/root n = 4/ root 36

SE=0.667

5. Step 5 – Compute Z Statistic

𝑍=74−72/0.667

Z≈3.00

6. Step 6 – Critical Value Method (Theoretical)

For 95% confidence level (two-tailed)

Zα/2=±1.96

Decision rule:

Reject 𝐻0 if ∣Z∣>1.96

Since

3.00>1.96

Reject the Null Hypothesis


7. Step 7 – p-Value Method (Practical)

From the Z-table:

For

Z=3.00

P(Z>3)=0.00135

Because it is a two-tailed test:

p=2×0.00135

p=0.0027

8. Step 8 – Compare With Significance Level

Confidence level = 95%

α=0.05

Compare:

p=0.0027

0.0027<0.05

Decision:

Reject 𝐻0

9. Final Conclusion

There is strong statistical evidence that the true average weight of City A is not 72 kg.

The sample suggests the population mean may be significantly higher than 72 kg.

At the 95% confidence level, there is sufficient evidence to reject the null hypothesis, indicating that the population mean weight is significantly different from 72 kg.​


```python
from scipy.stats import norm
import numpy as np

z=(74-72)/(4/np.sqrt(36))
p=2*(1-norm.cdf(abs(z)))

print(z,p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given values
mu = 72
sigma = 4
x_bar = 74
n = 36

# Z statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))

# x range for normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Standard Normal Distribution")

# Shade p-value region
plt.fill_between(x, y, where=(x >= z), color="red", alpha=0.4, label="p-value region")
plt.fill_between(x, y, where=(x <= -z), color="red", alpha=0.4)

# Mark Z value
plt.axvline(z, color="black", linestyle="--", label=f"Z = {z:.2f}")
plt.axvline(-z, color="black", linestyle="--")

plt.title("Z-Test Visualization")
plt.xlabel("Z value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
