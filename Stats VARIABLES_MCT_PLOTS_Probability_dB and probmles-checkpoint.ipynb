In statistics, Measures of Central Tendency (MCT) are used to represent the central or typical value of a dataset.

Main MCTs:

Mean

Median

Mode

But not all of them work for every type of variable. It depends on the data type.

Nominal → Mode only       data=pd.Series(Object ex[Gender, color,city,Blood Group])

Ordinal → Median, Mode    size=pd.Categorical(object ex[education level, satisfaction, shirt size],categories=[],ordered=True)

Numeric (Discrete/Continuous) → Mean, Median, Mode |numpy.random.randint(range, count), numpy.random.uniform(range), numpy.random.normal(range,count)

Measures of central tendency depend on the type of variable.

1. For nominal data, only mode is meaningful.
2. For ordinal data, median and mode can be used.
3. For numerical data (discrete or continuous), mean, median, and mode are all applicable.

```python
#1. Nominal Variables
import pandas as pd

colors=["Red", "Blue", "Blue", "Green", "Blue"]

mode_colors=pd.Series(colors).mode()[0]

print("Nomnal Variable:", colors)
print("Mode:",mode_colors)
```

```python
#2. Ordinal Variables
import pandas as pd
ratings=pd.Categorical(["Low","Medium","High","Low","High","Medium","High","Medium","High","Low"],categories=["Low","Medium","High"],ordered=True)

rating_series=pd.Series(ratings)

print("ordinal varibale:", rating_series)
print("Median:",rating_series.sort_values().iloc[len(rating_series)//2])
print("Mode",rating_series.mode()[0])

#Note: Sort the ordinal values when we going with the Median, Mode
```

##### 3. Normal Distribution (User Session Time)
Problem

Average session time on a website is 10 minutes with standard deviation 2 minutes.
Find the probability a session lasts less than 12 minutes.

This uses Normal distribution.


Z=X-mue/std


```python
from scipy.stats import norm

mean = 10
std = 2

prob = norm.cdf(12, mean, std)

print("Probability session < 12 minutes:", prob)
```

##### 4. Exponential Distribution (Customer Waiting Time)
Problem

Customers arrive at a support chat every 5 minutes on average.
What is the probability the next customer arrives after 10 minutes?

This uses Exponential distribution.

```python
from scipy.stats import expon

scale = 5

prob = 1 - expon.cdf(10, scale=scale)

print("Probability waiting > 10 minutes:", prob)
```

##### 5. A/B Testing (Beta Distribution)
Problem

Two versions of a webpage are tested:
| Version | Clicks | Views |
| ------- | ------ | ----- |
| A       | 40     | 200   |
| B       | 55     | 200   |


Which page likely performs better?

We model click probability using Beta distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

a_clicks, a_views = 40, 200
b_clicks, b_views = 55, 200

a_dist = beta(a_clicks+1, a_views-a_clicks+1)
b_dist = beta(b_clicks+1, b_views-b_clicks+1)

x = np.linspace(0,1,100)

plt.plot(x, a_dist.pdf(x), label="Version A")
plt.plot(x, b_dist.pdf(x), label="Version B")

plt.legend()
plt.title("A/B Testing using Beta Distribution")
plt.show()
```

##### 6. Simulation (Monte Carlo Method)

Data scientists often simulate events instead of solving formulas.

Example: simulate coin toss probability.

```python
import numpy as np

tosses = np.random.choice(["H","T"], size=10000)

prob_heads = np.mean(tosses == "H")

print("Estimated probability of heads:", prob_heads)
```
