```python
# Chebyshevs Inequality
#1. Take Income distribution usually skewded to the right. We can use Chebyshevs inequality to find out the percentage of people who earn more than 100k per year.
```

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2. Take data of income distribution.
data=np.random.exponential(scale=50, size=100000)


#Visualize the data
sns.histplot(data, bins=100, kde=True)
plt.show()
```

```python
#3. compute mean and standard deviation
mean=np.mean(data)
std=np.std(data)
print("Mean:", mean)
print("Standard Deviation:", std)

```

```python
#4. 
k=2
#5. Chebyshev range
lower_bound=mean-k*std
upper_bound=mean+k*std
print("Chebyshev Range:", lower_bound, upper_bound)
```

```python
#6. Actaul data inside the range
inside=np.mean((data>=lower_bound) & (data<=upper_bound))
print("Actual percentage of data inside the range:", inside*100)
```

```python
#Visualize the range
sns.histplot(data, bins=100, kde=True)
plt.axvline(lower_bound, color='red', linestyle='--', label='Lower Bound')
plt.axvline(upper_bound, color='green', linestyle='--', label='Upper    Bound')
plt.legend()
plt.show()
```

#Scenario (Real-world)

A company tracks delivery times (minutes).

Assume:

Mean delivery time = 30 minutes

Standard deviation = 5 minutes

We want to know how much data lies within 2 standard deviations (20–40 minutes).

Gaussian → precise estimate if data is normal

Chebyshev → safe minimum guarantee for any distribution

```python
#Compare Gaussian distribution with Chebyshev's inequality
import numpy as np

# simulate normal delivery times
data = np.random.normal(loc=30, scale=5, size=10000)

mean = np.mean(data)
std = np.std(data)

lower = mean - 2*std
upper = mean + 2*std

# actual percentage within 2 std
inside = np.mean((data >= lower) & (data <= upper))

print("Actual percentage within 2 std:", inside*100, "%")
```

```python
#Compare Gaussian distribution with Chebyshev's inequality
import numpy as np

# simulate normal delivery times
data = np.random.exponential(scale=5, size=10000)

mean = np.mean(data)
std = np.std(data)
#k=2
lower = mean - 2*std
upper = mean + 2*std

# actual percentage within 2 std
inside = np.mean((data >= lower) & (data <= upper))

print("Actual percentage within 2 std:", inside*100, "%")
```

```python
#visualize the range
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data, bins=40, kde=True)

plt.axvline(lower, color='red', linestyle='--')
plt.axvline(upper, color='green', linestyle='--')

plt.title("Delivery Time Distribution (±2σ Range)")
plt.xlabel("Delivery Time (minutes)")
plt.show()
```

Both methods help estimate how much data lies within a certain number of standard deviations (σ) from the mean (μ).
1. Gaussian (Empirical) Rule

Works only when data follows a normal distribution.

Gives approximate percentages of data within standard deviations.

2. Chebyshev’s Inequality

Works for any distribution (normal or skewed).

Gives a minimum guaranteed percentage of data within k standard deviations.
