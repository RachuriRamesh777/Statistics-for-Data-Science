#Central limit theorem: Definition:


The Central Limit Theorem states that when we take many random samples from a population and calculate their sample means, the distribution of those means will approach a normal distribution as the sample size becomes large (usually 
𝑛
≥
30
n≥30), regardless of the original population distribution.

```python
# STEP 1:
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Step 2: Consider Non Normal/ skewed data population:
population=np.random.exponential(scale=2, size=100000)
#confirm population in not normal by plot
import seaborn as sns

sns.histplot(population, bins=50, kde=True)
plt.title("Skewed Population Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Step 3: Take size of the n samples=10000, and each sample=30 values
sample_means=[]
for i in range(1000):
    sample=np.random.choice(population,size=30)
    sample_means.append(np.mean(sample))
print(sample_means)
```

```python
import seaborn as sns

sns.histplot(sample_means, bins=30, kde=True)

plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()
```

Note: With the help of sample_means we take it refernce for population
And Because mean of sample_means= population mean, std=sigma/rot n
we compute statistical inferences with sample_means as refence for population
1. Poplation mean using sample mean
2. Std or standerd error of population using sample_means
3. confidence intervals
4. perform Hypothesis testing

Make data-driven decisions, Even though the original data was skewed.

```python
#1. POPLATION MEAN USING SAMPLE MEAN
mean=np.mean(sample_means)
print(mean)
#2. Standerd error/ std of the population
std=np.std(sample_means)
print(std)
#3. Build confidance intervals

lower=mean-1.93*std
upper=mean+1.93*std
print(lower, upper)

#So we are 95%confidentto tell that true population mean lies bitween lower and upper confidence intervals
#4 Hypothesis testing: wether
#H0: Poplation Mean=2, H1 population mean not=2

from scipy import stats
stats.ttest_1samp(sample_means,2)
```

```python
#resslt we pass to show H0 =2 and failed to show H1=2, as the P value>0.05.
# So the Result: Population Mean is equal to 2.
```

# Clt with Delivery Time Analysis:
A food delivery company wants to know:

Is the average delivery time greater than 30 minutes?

The delivery times are skewed (traffic, distance, etc.), so the population is not normal.

```python
#1: Simaluate delivery Time Data which is not normal, skewed
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Deliver time data population
population=np.random.exponential(scale=30,size=100000)

sns.histplot(population, kde=True)
plt.title("Delivery Time Population (Skewed)")
plt.show()
```

```python
#Result: Delivery Time is skewed
#2. Take many samples for applying CLT 
sample_means=[]

for i in range(1000):
    sample=np.random.choice(population, size=40)
    sample_means.append(np.mean(sample))
```

```python
#3. Clt result
sns.histplot(sample_means, bins=30, kde=True)
plt.title("Sampling Distribution of Mean (CLT)")
plt.xlabel("Average Delivery Time")
plt.show()
```

```python
# so sample mean ditribution become Normal apply
#4 statistial Inference (confidence interval)
mean=np.mean(sample_means)
std=np.std(sample_means)

lower =mean-1.96*std
upper=mean+1.96*std

print("Mean:", mean)
print("95% Confidence Interval:", lower, upper)
```

#### We are 95% confident the true average delivery time is between 20 and 39.388minutes.
Prove with Hypothesis Testing 

```python
#5.Hypothesis test
#H0 DELIVERY TIME IS=30 MINUTES
#H1 not=30 mins

#Prove

from scipy import stats
stats.ttest_1samp(sample_means,30)
```

#### So the delivery time of the company is 30. we passed to prove H0 and failed to prove H1 as p value 0.400>0.05 so 
#the claiming of the company delivery time is 30 minutes True

The company can decide

If delivery time is acceptable

Whether to improve logistics

Whether to add more delivery agents

### CLT allows us to analyze large real-world skewed data (delivery time, income, sales) using normal distribution tools like confidence intervals and hypothesis testing.
