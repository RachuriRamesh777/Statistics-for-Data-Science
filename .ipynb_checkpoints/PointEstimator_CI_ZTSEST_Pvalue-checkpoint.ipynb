# Q3. Decision Making: one company claims avg daily sales are 5000
sample=4800, std s=800, n=64
Test for ci95% is the claime tru?
Usinh Hypothesis testing


### 1. Using Ci approach

```python
#1. Define H0: =5000, H1 not=5000
s_m=sample_mean= 4800
se=standard_error=800/np.sqrt(64)
#Zscore for 95% ci is 1.96 using Z table

z=1.96
lower=s_m-z*se
upper=s_m+z*se


print("1. pop mean Mue", s_m)
print("2. CI")
print("Lower: ",lower)
print("Upper: ", upper)
```

### Company claims are 5000 not lies in CI so the HO is not True and reject it

### 2. Using Z test aproach  as std, sample size>30 we use z test rather then Ttest in case of no std

```python
import numpy as np
from scipy import stats

# Given data
s_m = 4800
pop_mean = 5000
se = 800 / np.sqrt(64)

# Z-score
z = (s_m - pop_mean) / se
print("Z-score:", z)

# p-value (two-tailed)
p_val = 2 * (1 - stats.norm.cdf(abs(z)))
print("p-value:", p_val)
```

### Result: As the p value 0.045 is less than 0.05 significant value so H0 is rejected so avaerage daily sales 5000is not ture. 
In actual daily sales are lies in bitween 4604 to 4996 only in sample of 64 days

## Final Conclusion

- Point estimate gives a single value (sample mean)
- Confidence interval provides a range of plausible values
- Hypothesis testing validates claims using statistical evidence
- p-value confirms statistical significance

Hence, the company’s claim of ₹5000 is not supported by data.

### A/B Testing Case Study (Real Company Scenario)

This is where everything you learned (CI + p-value + hypothesis testing) becomes business decision-making.

Problem (Real Scenario)

An e-commerce company tests a new website design.

Old design (A) → average sales = ₹5000

New design (B) → sample data:

Sample mean = ₹5300

Std deviation = ₹800

Sample size = 64

👉 Question:

Should the company switch to the new design?

```python
import numpy as np
from scipy import stats

# Given
sample_mean = 5300
pop_mean = 5000
std_dev = 800
n = 64

# Standard error
se = std_dev / np.sqrt(n)

# Z-score
z = (sample_mean - pop_mean) / se

# Confidence Interval (95%)
z_critical = 1.96
lower = sample_mean - z_critical * se
upper = sample_mean + z_critical * se

# p-value (one-tailed)
p_value = 1 - stats.norm.cdf(z)

print("1. Z-score:", z)
print("2. p-value:", p_value)
print("3. CI:", lower, "to", upper)

# Decision
if p_value < 0.05:
    print("Decision: Reject H0")
    print("Conclusion: New design significantly improves sales ✅")
else:
    print("Decision: Fail to reject H0")
    print("Conclusion: No significant improvement ❌")
```
